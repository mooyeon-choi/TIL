```python
# 아래에 코드를 작성해주세요.
import random
class Pikachu:
    
    def __init__(self, name):
        self.name = name+'츄'
        self.level = 5
        self.hp = self.level * random.randint(15, 20)
        self.exp = 0
        
    def bark(self):
        return 'pikachu'
    
    def attack(self, target):
        
        if random.randint(0, 1):
            print(f'{self.name}의 몸통박치기!')
            if random.randint(0, 1):
                target.hp -= self.level * random.randint(2, 5)
                print(f'{target.name}의 체력이 {target.hp}가 되었다....!')
            else:
                print('공격에 실패했다.')
        else:
            print(f'{self.name}의 천만볼트!')
            if random.randint(0, 1):
                target.hp -= self.level * random.randint(4, 7)
                print(f'{target.name}의 체력이 {target.hp}가 되었다....!')
            else:
                print('공격에 실패했다.')
        if target.hp <= 0:
            self.exp += self.level * 15
            if self.exp >= 100:
                self.level += 1
                self.exp = 0
                self.hp = self.level * random.randint(15, 20)
            print(f'{target.name}: 으앙 주금')
        return target.hp
    
    def __repr__(self):
        return f'{self.name} hp[{self.hp}]: lv.{self.level} exp.{self.exp}/100'
    def __del__(self):
        print(f'{self.name}은(는) 버려졌다.')
```

```python
turn = 0
p1_name = input('플레이어 1의 이름을 정해주세요.')
p2_name = input('플레이어 2의 이름을 정해주세요.')
p1 = Pikachu(p1_name)
p2 = Pikachu(p2_name)
print('===========게임이 시작되었습니다================')
while True:
    if p1.hp <= 0:
        p1_name = input('플레이어 1의 이름을 정해주세요.')
        p1 = Pikachu(p1_name)
    elif p2.hp <= 0:
        p2_name = input('플레이어 2의 이름을 정해주세요.')
        p2 = Pikachu(p2_name)
    while True:
        turn += 1
        print(f'=============={turn}================')
        if turn % 2:
            p1.attack(p2)
            print(p1)
            print(p2)
        else:
            p2.attack(p1)
            print(p1)
            print(p2)
        if p1.hp <= 0:
            p2.hp = p2.level * random.randint(15, 20)
            break
        elif p2.hp <= 0:
            p1.hp = p1.level * random.randint(15, 20)
            break
```
